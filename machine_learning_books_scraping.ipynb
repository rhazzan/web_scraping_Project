```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "https://libgen.li/index.php?req=machine+learning&columns%5B%5D=t&objects%5B%5D=f&topics%5B%5D=l&res=100&covers=on&filesuns=all"

response = req.get(url)
soup = bs(response.content,"html")
print(soup.prettify())
```

```python
table = soup.find("table",{"class": "table"})
table
```

```python
headers = table.find("thead")
headers = [col.text.strip().replace("↕","").strip().replace("\r","").replace("\n","") for col in headers.find_all("th")][1:]
headers[0] = "Name"
headers
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
image_src = []
title_data = []
body = soup.find("tbody")
md5 = []
for row in body.find_all("tr"):
    row_data = [td for td in row.find_all("td")]
    image_src.append(row_data[0].find("img")["src"])
    if row_data[1].find("a").text == "":
       title_data.append(row_data[1].find("b").text)
    else:
        title_data.append(row_data[1].find("a").text)
    other_columns = [td.text for td in row.find_all("td")[1:]]
    md5.append(row_data[9].find("a")["href"].split("=")[-1].split("/")[-1])

    lenght = len(df)

    df.loc[lenght] = other_columns
df
```

```python
for i,item in enumerate(title_data):
    df["Name"][i] = item

df["Name"]
```

```python
for i,item in enumerate(md5):
    df["Mirrors"][i] = "https://annas-archive.gl/md5/" + item
df["Author(s)"]= df["Author(s)"].str.replace(";",",")
df.loc[0]
```

```python
df.drop_duplicates(inplace = True,)
df["Name"].value_counts()
```

```python
df.to_excel("Machine_learning_textbooks.xlsx",index=False)
```
